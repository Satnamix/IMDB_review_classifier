```python
## word Embedding Representation
from tensorflow.keras.preprocessing.text import one_hot # type:ignore
from tensorflow.keras.layers import Embedding # type:ignore
# from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences # type:ignore
from tensorflow.keras.models import Sequential # type:ignore
import numpy as np
```

```python
### sentences
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
voc_size = 10000
```

```python
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,sent_length)
embedded_docs
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
